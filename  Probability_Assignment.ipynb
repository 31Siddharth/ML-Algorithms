```python
'''
Your friends comes up with a fair coin and proposes the following. You will toss the coin untill heads comes up. For each tails you will pay your friend 10$, But if a head comes up your friend will pay you 15$

The sample space of such an experiment looks like this {H, TH, TTH, .... TTTTTH ....}. As soon as a head comes up you stop tossing and your friend pays you 100$. You need to pay your friend 10$ for each tails that came up. So to make a profit you need to get head as quickly as possible.

You can use the simulate a fair coin like we did above. Now you use power of simulation.

Write a simulation logic and a trial logic with which you can decide if you should take this bet or not.

What if your friend offered you 100$ once head comes up? Will you take the bet?

Write a simulation function to decide.'''
```

```python
import random
def coin_toss():
  if random.random() <= 0.5:
    return 0
  return 1

def game():
  give = 0
  recieve = 0
  #tosses = []
  while True:
    toss = coin_toss()
    #tosses.append(toss)
    if(toss == 0):
      give += 10 
      continue
    elif (toss == 1):
      recieve += 15
      break
  #print(tosses)
  return recieve - give

def simulate(n):
  res = []
  for i in range(n):
    res.append(game())
  return sum(res)
```

```python
import random
def dice_roll():
  dice1 = random.randint(1,6)
  dice2 = random.randint(1,6)
  return dice1 + dice2

def simulate(n):
  rolls = []
  for i in range(n):
    wins = 0
    while dice_roll() != 12:
      wins -= 5

    wins += 100
    rolls.append(wins)
    
  print(sum(rolls)/n)
```

```python
#Spam Classifier
import pandas as pd

sms_spam = pd.read_csv('SMSSpamCollection', sep='\t',
header=None, names=['Label', 'SMS'])
sms_spam
```

```python
#Using bayes theorem P(spam|sms) = P(sms|spam)*P(spam)/(P(sms|spam)*P(spam) + P(sms|ham)*P(ham))
p_spam = sms_spam[sms_spam['Label'] == 'spam'].Label.count()/sms_spam['Label'].count()
p_ham = sms_spam[sms_spam['Label'] == 'ham'].Label.count()/sms_spam['Label'].count()
sms_spam['Label'].value_counts(normalize = True)
```

```python
# Randomize the dataset
data_randomized = sms_spam.sample(frac=1, random_state=1)

# Calculate index for split
training_test_index = round(len(data_randomized) * 0.8)

# Split into training and test sets
training_set = data_randomized[:training_test_index].reset_index(drop=True)
test_set = data_randomized[training_test_index:].reset_index(drop=True)

print(training_set.shape)
print(test_set.shape)
```

```python
# After cleaning
training_set['SMS'] = training_set['SMS'].str.replace(
   '\W', ' ') # Removes punctuation #\W matches characters which is not a word character [^a-zA-Z0-9_]
training_set['SMS'] = training_set['SMS'].str.lower()
training_set
```

```python
#Creating a vocabulary
vocabulary = {}
for i in training_set['SMS']:
    words = i.split(' ')
    for word in words:
        if word not in vocabulary.keys():
            vocabulary[word] = 1
        else:
            vocabulary[word] +=1
len(vocabulary.keys())
```

```python
#creating vocabulary as a list
training_set['SMS'] = training_set['SMS'].str.split()

vocabulary = []
for sms in training_set['SMS']:
   for word in sms:
      vocabulary.append(word)

vocabulary = list(set(vocabulary))    
```

```python
word_counts_per_sms = {unique_word: [0] * len(training_set['SMS']) for unique_word in vocabulary}

for index,sms in enumerate(training_set['SMS']):
    for word in sms:
        word_counts_per_sms[word][index] = 1
```

```python
word_counts = pd.DataFrame(word_counts_per_sms)
word_counts
```

```python
training_set_clean = pd.concat([training_set, word_counts], axis=1)
training_set_clean
```

```python
# Isolating spam and ham messages first
spam_messages = training_set_clean[training_set_clean['Label'] == 'spam']
ham_messages = training_set_clean[training_set_clean['Label'] == 'ham']

# P(Spam) and P(Ham) in training set
p_spam = len(spam_messages) / len(training_set_clean)
p_ham = len(ham_messages) / len(training_set_clean)

# N_Spam
n_spam = len(spam_messages)

# N_Ham

n_ham = len(ham_messages)

# N_Vocabulary
n_vocabulary = len(vocabulary)

# Laplace smoothing
alpha = 1
```

```python
''' 𝑃(𝑤𝑖|𝑆𝑝𝑎𝑚)=𝑁𝑖|𝑆𝑝𝑎𝑚+𝛼/𝑁𝑠𝑝𝑎𝑚+𝛼.𝑁𝑣 

    𝑃(𝑤𝑖|𝐻𝑎𝑚)=𝑁𝑖|𝐻𝑎𝑚+𝛼/𝑁ℎ𝑎𝑚+𝛼.𝑁𝑣  '''

# Initiate parameters
# parameters_spam[word] will give probability of word occuring in spam email
parameters_spam = {unique_word:0 for unique_word in vocabulary}
# parameters_ham[word] will give probability of word occuring in ham email
parameters_ham = {unique_word:0 for unique_word in vocabulary}

# Calculate parameters
for word in vocabulary:
    # Ni|Spam + alpha 
    n_word_given_spam = spam_messages[word].sum() # spam_messages already defined
    # P(wi|Spam)
    p_word_given_spam = (n_word_given_spam + alpha) / (n_spam + alpha*n_vocabulary)
    parameters_spam[word] = p_word_given_spam

    # Ni|Ham + alpha 
    n_word_given_ham = ham_messages[word].sum() # ham_messages already defined
    # P(wi|Ham)
    p_word_given_ham = (n_word_given_ham + alpha) / (n_ham + alpha*n_vocabulary)
    parameters_ham[word] = p_word_given_ham
```

```python
import re

# message : string
def spam_detector(message):
   
    message = re.sub('\W','', message)
    message = message.lower().split()

    p_spam_given_message = p_spam
    p_ham_given_message = p_ham

    for word in message:
        if word in parameters_spam:
            # p(Spam|word) = p(spam) * p(word|spam) = prior * likelihood
            p_spam_given_message *= parameters_spam[word]

        if word in parameters_ham:
            p_ham_given_message *= parameters_ham[word]


    p_given_message = p_spam_given_message + p_ham_given_message

    # p(Spam|word) = (p(spam) * p(word|spam)) / p(word) 
    p_spam_given_message = p_spam_given_message/p_given_message

    p_ham_given_message = p_ham_given_message/p_given_message

    #print('P(Spam|message):', p_spam_given_message)
    #print('P(Ham|message):', p_ham_given_message)

    if p_ham_given_message > p_spam_given_message:
        return 'ham'
    elif p_ham_given_message < p_spam_given_message:
        return 'spam'
    else:
        return 'Equal proabilities, needs human classification'
```

```python
spam_detector('WINNER!! This is the secret code to unlock the money: C3421.')
```

```python
test_set['Predicted'] = test_set['SMS'].apply(spam_detector)
test_set[test_set['Predicted'] == 'ham']
```

```python
correct = 0
total = test_set.shape[0]

for row in test_set.iterrows():
    row = row[1]
    if row['Label'] == row['Predicted']:
        correct += 1

print('Correct:', correct)
print('Incorrect:', total - correct)
print('Accuracy:', correct/total)
```
